```python
import cv2
import os

# Set the path to the directory containing the images
image_path = "D:\\ENGINE\\Detection101\\processed"

# Get a list of all the image files in the directory
image_files = sorted([os.path.join(image_path, f) for f in os.listdir(image_path)])

# Set the height of the final output image
height = cv2.imread(image_files[0]).shape[0]

# Initialize an empty list to hold the resized images
resized_images = []

# Loop through each image and resize it to the same height
for image_file in image_files:
    image = cv2.imread(image_file)
    width = int((height / image.shape[0]) * image.shape[1])
    resized_image = cv2.resize(image, (width, height))
    resized_images.append(resized_image)

# Combine the resized images horizontally
combined_image = cv2.hconcat(resized_images)

# Set the size of the sliding box
box_size = (1207, 1207)

# Initialize the video writer
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
output_path = "D:\\ENGINE\\Detection101\\WithboxResult3.mp4"
out = cv2.VideoWriter(output_path, fourcc, 20.0, (box_size[0], box_size[1]))

# Define the bounding box parameters
box_width = 1207
box_height = 1000
box_color = (0, 0, 255)  # Red

# Loop through each frame in the combined image
for x in range(0, combined_image.shape[1] - box_size[0], 10):
    # Extract the current frame
    frame = combined_image[:, x:x+box_size[0]]

    # Resize the frame to the size of the sliding box
    frame = cv2.resize(frame, box_size)

    # Add the bounding box to the frame
    x_offset = int((box_size[0] - box_width) / 2)
    y_offset = int((box_size[1] - box_height) / 2)
    cv2.rectangle(frame, (x_offset, y_offset), (x_offset+box_width, y_offset+box_height), box_color, thickness=2)

    # Write the frame to the output video
    out.write(frame)

# Release the video writer and destroy any remaining windows
out.release()
cv2.destroyAllWindows()
```
